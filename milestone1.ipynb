```python
# Week 1: Data Foundation
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "flight.csv"
df = pd.read_csv(file_path, low_memory=False)

print("First 5 records:")
display(df.head())

print("\nDataset Info:")
df.info()
print("\nSummary Statistics:")
display(df.describe(include='all'))

# Drop duplicates
df.drop_duplicates(inplace=True)

# Check null values
print("\nMissing Values:")
print(df.isnull().sum()[df.isnull().sum() > 0])

# Sampling for large dataset
sample_df = df.sample(100000, random_state=42) if len(df) > 100000 else df.copy()
print(f"\nSample size used: {len(sample_df)}")
```

```python
# Week 2: Preprocessing & Feature Engineering

#  Handle nulls in delay and cancellation columns
for col in ['DEP_DELAY', 'ARR_DELAY', 'CANCELLED',
            'DELAY_DUE_CARRIER','DELAY_DUE_WEATHER',
            'DELAY_DUE_NAS','DELAY_DUE_SECURITY','DELAY_DUE_LATE_AIRCRAFT']:
    if col in sample_df.columns:
        sample_df[col] = sample_df[col].fillna(0)

#  Convert FL_DATE to datetime & create derived features
if 'FL_DATE' in sample_df.columns:
    sample_df['FL_DATE'] = pd.to_datetime(sample_df['FL_DATE'], errors='coerce')
    sample_df['Month'] = sample_df['FL_DATE'].dt.month
    sample_df['DayOfWeek'] = sample_df['FL_DATE'].dt.dayofweek  # 0=Mon, 6=Sun

#  Extract Hour from CRS_DEP_TIME
if 'CRS_DEP_TIME' in sample_df.columns:
    sample_df['CRS_DEP_TIME'] = sample_df['CRS_DEP_TIME'].astype(str).str.zfill(4)
    sample_df['Hour'] = sample_df['CRS_DEP_TIME'].str[:2].astype(int)

#  Create Route feature
if 'ORIGIN' in sample_df.columns and 'DEST' in sample_df.columns:
    sample_df['Route'] = sample_df['ORIGIN'] + "  " + sample_df['DEST']

#  Format other datetime columns (DEP_TIME, ARR_TIME, WHEELS_OFF, WHEELS_ON)
time_cols = ['DEP_TIME', 'ARR_TIME', 'WHEELS_OFF', 'WHEELS_ON']
for col in time_cols:
    if col in sample_df.columns:
        sample_df[col] = sample_df[col].astype(str).str.zfill(4)  # HHMM
        sample_df[col] = pd.to_datetime(
            sample_df['FL_DATE'].astype(str) + ' ' + sample_df[col],
            format='%Y-%m-%d %H%M', errors='coerce'
        )

#  Save cleaned dataset
sample_df.to_csv("flights_cleaned.csv", index=False)
print("Preprocessed data saved as flights_cleaned.csv")
```
